**From TF to TFLite**

```python
import tensorflow.lite as tflite
```

```python
converter = tflite.TFLiteConverter.from_keras_model(model)

converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS, tf.lite.OpsSet.SELECT_TF_OPS]
converter._experimental_lower_tensor_list_ops = False
```

```python
tflite_model = converter.convert()
```

```python
with open('mugen-model.tflite', 'wb') as f_out:
    f_out.write(tflite_model)
```

**predicting with TFLite**

```python
import tflite_runtime.interpreter as tflite
```

```python
interpreter = tflite.Interpreter(model_path='mugen-model.tflite')
interpreter.allocate_tensors()

input_index = interpreter.get_input_details()[0]['index']
output_index = interpreter.get_output_details()[0]['index']
```

```python
X_reshaped = X_reshaped.astype('float32')
```

```python
interpreter.set_tensor(input_index, X_reshaped)
interpreter.invoke()
preds = interpreter.get_tensor(output_index)
```

```python
genre_preds = dict(zip(genres, preds[0]))

max_genre = max(genre_preds, key=genre_preds.get)
max_value = genre_preds[max_genre]

max_genre, max_value
```

```python
def predict(X):
    
    preds = model.predict(X)
    #float_predictions = preds[0].tolist()

    genre_preds = dict(zip(genres, preds[0]))

    max_genre = max(genre_preds, key=genre_preds.get)
    max_value = genre_preds[max_genre]

    return genre_preds, max_genre, max_value
```
